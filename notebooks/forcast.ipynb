```python
zebra_data = pd.read_csv('zebra.csv')
```

```python
zebra_data.reset_index(inplace=True)
```

```python
df = pd.DataFrame()
df['ds'] = pd.to_datetime(zebra_data['Order Date'])
df['y'] = zebra_data['Sale Price'] 
```

```python
prophet = Prophet()
prophet.fit(df)
future = prophet.make_future_dataframe(periods=12 * 4, freq='D')
forecast = prophet.predict(future)
fig = prophet.plot(forecast)
#a = add_changepoints_to_plot(fig.gca(), prophet, forecast)
```
